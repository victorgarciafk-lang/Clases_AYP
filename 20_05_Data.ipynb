```python
import matplotlib.pyplot as plt
```

```python
df_line = data.groupby("bedrooms")["price"].mean()

plt.figure(figsize=(8,5))
plt.plot(df_line.index, df_line.values)
plt.title("Precio promedio según número de habitaciones")
plt.xlabel("Número de habitaciones")
plt.ylabel("Precio promedio")
plt.tight_layout()
plt.show()
```

```python
df_line = data.groupby('bedrooms')['price'].mean()

plt.figure(figsize=(8,5))
plt.plot(df_line.index, df_line.values)
plt.title("Precio promedio según número de habitaciones")
plt.xlabel("Número de habitaciones")
plt.ylabel("Precio promedio")
plt.tight_layout()
plt.show()

plt.figure(figsize=(8,5))
data['bedrooms'].value_counts().plot(kind='bar')
plt.title("Distribución de número de habitaciones")
plt.xlabel("Habitaciones")
plt.ylabel("Cantidad de viviendas")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(data['sqft_living'], data['price'], s=10, alpha=0.3)
plt.title("Precio vs Área habitable")
plt.xlabel("sqft living")
plt.ylabel("Precio")
plt.tight_layout()
plt.show()
```

```python
renov = data["renovated_flag"].value_counts()
etiquetas = ["No renovada", "Renovada"]
renov.index = etiquetas

plt.figure(figsize=(6,6))
plt.pie(renov.values, labels=renov.index, autopct='%1.1f%%')
plt.title("Porcentaje de casas renovadas")
plt.tight_layout()
plt.show()
```
