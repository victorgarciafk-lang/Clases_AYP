```python
indice = ['gerente', 'supervisor', 'vendedor', 'cajero']
personal = pd.DataFrame({'nombres':('Juan Pérez',
                                    'María Sánchez',
                                    'Jorge Vargas',
                                    'Rodrigo Martínez'),'fechas':(datetime(1995,12,21),
                      datetime(1989,1,13),
                      datetime(1992,9,14),
                      datetime(1993,7,8)),
            'saldo': (2500,
                      5345,
                      np.nan,
                      11323.2),
            'al corriente':(True,
                            True,
                            False,
                            True)}, index=indice)
personal
```

```python
personal.shape
```

```python
personal.index#Para saber los valores de los indices
```

```python
personal.columns#Para saber los nombres de las columnas
```

```python
personal.reset_index()
```

```python
p=personal.set_index("nombres")#Para cambiar el indice por una de los columnas en este caso nombres
p
```

```python
indice_1 = ("lunes", "martes", "miércoles")
marco_1 = pd.DataFrame({'Centro':[1520,1640,1043],
                       'Sur':[1422,1673,1534],
                      'Oriente':[1021,1073,1100]},
                       index=indice_1)

indice_2 = ("jueves", "viernes", "sábado")
marco_2 = pd.DataFrame({'Centro':[1321,1459,1875],
                       'Sur':[1622,1841,1920],
                      'Oriente':[1500,1432,1491]},
                      index=indice_2)
```

```python
marco_2
```

```python
marco_1
```

```python
pd.concat([marco_1,marco_2],axis=0)
```
